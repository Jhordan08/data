```
import matplotlib.pyplot as plt
import numpy as np

# cargar los datos del archivo .dat
datos = np.loadtxt('p204-1.txt')

# extraer las columnas de datos
x = datos[:, 0]
y = datos[:, 1]
z = datos[:, 2]
color = datos[:, 2]

# crear la figura y los ejes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# agregar los puntos a la gráfica
puntos = ax.scatter(x, y, z, c=color)

# agregar la barra de color
barra_color = fig.colorbar(puntos)

# agregar una etiqueta a la barra de color
barra_color.set_label('amplitud optima')

# agregar etiquetas a los ejes
ax.set_xlabel('a')
ax.set_ylabel('beta')
ax.set_zlabel('omega optimo')

# mostrar la gráfica
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo .dat
datos = np.loadtxt('lib.txt')

# Extraer las columnas de datos
x = datos[:,0]
y = datos[:,1]
color = datos[:,3]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar los datos con colores
puntos = ax.scatter(x, y, c=color)

# Crear una barra de colores
barra_colores = fig.colorbar(puntos, ax=ax)
barra_colores.set_label('amplitud optima')

# Añadir títulos y etiquetas de los ejes
ax.set_title('Gráfico con amplitudes con colores')
ax.set_xlabel('a')
ax.set_ylabel('beta')

# Mostrar la figura
plt.show()
```
